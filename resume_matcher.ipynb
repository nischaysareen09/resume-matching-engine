```python
import math

# ─────────────────────────────────────────────
# SKILL ALIASES
# ─────────────────────────────────────────────
SKILL_ALIASES = {
    "python": "python", "pyhton": "python",
    "java": "java",
    "javascript": "javascript", "javascrpit": "javascript", "js": "javascript",
    "typescript": "typescript", "typescrpit": "typescript",
    "c++": "cpp", "cpp": "cpp",
    "r": "r",
    "kotlin": "kotlin",
    "machinelearning": "machine_learning", "machine learning": "machine_learning",
    "ml": "machine_learning", "sklearn": "machine_learning",
    "deeplearning": "deep_learning", "deep learning": "deep_learning",
    "deep-learning": "deep_learning",
    "tensorflow": "tensorflow", "pytorch": "pytorch", "keras": "keras",
    "nlp": "nlp", "bert": "bert", "xgboost": "xgboost",
    "feature engineering": "feature_engineering",
    "statistics": "statistics", "stats": "statistics",
    "regression": "regression", "clustering": "clustering",
    "data-viz": "data_visualization", "data visualization": "data_visualization",
    "data viz": "data_visualization", "matplotlib": "data_visualization",
    "tableau": "data_visualization", "power-bi": "data_visualization",
    "power bi": "data_visualization", "powerbi": "data_visualization",
    "pandas": "pandas", "numpy": "numpy",
    "react": "react", "reacts": "react", "reactjs": "react",
    "vue": "vue", "vue.js": "vue", "vuejs": "vue",
    "redux": "redux", "tailwind": "tailwind",
    "html/css": "html_css", "html css": "html_css",
    "html": "html_css", "css": "html_css",
    "jest": "jest", "graphql": "graphql",
    "node.js": "nodejs", "nodejs": "nodejs", "node js": "nodejs",
    "flask": "flask",
    "spring boot": "spring_boot", "springboot": "spring_boot",
    "rest api": "rest_api", "rest": "rest_api", "restapi": "rest_api",
    "microservices": "microservices",
    "sql": "sql", "mysql": "mysql", "mysq": "mysql",
    "postgresql": "postgresql", "postgres": "postgresql",
    "mongodb": "mongodb", "redis": "redis",
    "docker": "docker",
    "kubernetes": "kubernetes", "kubernates": "kubernetes", "k8s": "kubernetes",
    "ci/cd": "ci_cd", "cicd": "ci_cd", "ci cd": "ci_cd",
    "aws": "aws",
    "android": "android", "firebase": "firebase",
    "algorithms": "algorithms", "algoritms": "algorithms",
    "data structure": "data_structures", "data structures": "data_structures",
    "competitive programming": "competitive_programming",
    "ui/ux": "ui_ux", "ui ux": "ui_ux", "figma": "figma",
}

# ─────────────────────────────────────────────
# RESUME DATASET
# ─────────────────────────────────────────────
resumes = [
    {"id": "01", "name": "Arjun Sharma",    "skills": "Pyhton, MachineLearning, SQL, pandas, numpy, Deep-learning"},
    {"id": "02", "name": "Priya Nair",      "skills": "JavaScrpit, Reacts, Node.JS, MongoDb, REST api, HTML/CSS"},
    {"id": "03", "name": "Rahul Gupta",     "skills": "Java, Spring Boot, MySql, Microservices, Docker, kubernates"},
    {"id": "04", "name": "Sneha Patel",     "skills": "Python, TensorFlow, Keras, NLP, BERT, data-viz, matplotlib"},
    {"id": "05", "name": "Vikram Singh",    "skills": "C++, Algoritms, Data Structure, competitive programming, python"},
    {"id": "06", "name": "Ananya Krishnan", "skills": "javascript, vue.js, python, flask, PostgreSQL, AWS, CI/CD"},
    {"id": "07", "name": "Karan Mehta",     "skills": "Python, Sklearn, XGboost, feature engineering, SQL, tableau"},
    {"id": "08", "name": "Deepika Rao",     "skills": "Java, Android, Kotlin, Firebase, REST, UI/UX, figma"},
    {"id": "09", "name": "Aditya Kumar",    "skills": "Reactjs, TypeScrpit, GraphQL, redux, tailwind, nodejs, jest"},
    {"id": "10", "name": "Meera Iyer",      "skills": "python, R, statistics, ML, regression, clustering, Power-BI"},
]

# ─────────────────────────────────────────────
# JOB DESCRIPTIONS
# ─────────────────────────────────────────────
job_descriptions = [
    {
        "id": "JD-1", "company": "Kakao", "role": "ML Engineer",
        "skills": ["python", "machine_learning", "deep_learning", "tensorflow",
                   "pytorch", "sql", "data_visualization", "nlp", "bert",
                   "feature_engineering", "statistics"]
    },
    {
        "id": "JD-2", "company": "Naver", "role": "Backend Engineer",
        "skills": ["java", "spring_boot", "mysql", "postgresql", "microservices",
                   "docker", "kubernetes", "rest_api", "ci_cd", "redis"]
    },
    {
        "id": "JD-3", "company": "Line", "role": "Frontend Engineer",
        "skills": ["javascript", "react", "vue", "typescript", "rest_api",
                   "html_css", "nodejs", "graphql", "redux", "jest", "aws"]
    },
]

# ─────────────────────────────────────────────
# STEP 1 - NORMALIZE SKILLS
# ─────────────────────────────────────────────
def normalize_skills(skill_string):
    tokens = [t.strip().lower() for t in skill_string.split(",")]
    # longest keys first so multi-word matches before single tokens
    sorted_keys = sorted(SKILL_ALIASES.keys(), key=lambda k: -len(k))
    normalized = []
    seen = set()
    for token in tokens:
        for alias in sorted_keys:
            if token == alias:
                canonical = SKILL_ALIASES[alias]
                if canonical not in seen:
                    normalized.append(canonical)
                    seen.add(canonical)
                break
    return normalized

# ─────────────────────────────────────────────
# STEP 2 - DEDUPLICATE SKILLS
# ─────────────────────────────────────────────
def deduplicate_skills(normalized_skills):
    seen = set()
    deduped = []
    for skill in normalized_skills:
        if skill not in seen:
            deduped.append(skill)
            seen.add(skill)
    return deduped

# ─────────────────────────────────────────────
# STEP 3 - BUILD VOCABULARY
# ─────────────────────────────────────────────
def construct_vocabulary(resumes):
    vocabulary = set()
    for resume in resumes:
        normalized = normalize_skills(resume["skills"])
        deduped = deduplicate_skills(normalized)
        vocabulary.update(deduped)
    return sorted(list(vocabulary))

# ─────────────────────────────────────────────
# STEP 4 - COMPUTE TF-IDF
# ─────────────────────────────────────────────
def compute_tfidf(resumes, vocabulary):
    # calculate idf once for all skills
    idf = {}
    for skill in vocabulary:
        df = sum(1 for r in resumes if skill in normalize_skills(r["skills"]))
        idf[skill] = math.log(10 / df) if df != 0 else 0

    tfidf_vectors = []
    for resume in resumes:
        skills = deduplicate_skills(normalize_skills(resume["skills"]))
        n = len(skills)
        tfidf_vector = []
        for skill in vocabulary:
            if skill in skills:
                tf = 1 / n
                tfidf_vector.append(tf * idf[skill])
            else:
                tfidf_vector.append(0)
        tfidf_vectors.append(tfidf_vector)
    return tfidf_vectors

# ─────────────────────────────────────────────
# STEP 5 - BUILD JD BINARY VECTORS
# ─────────────────────────────────────────────
def construct_binary_vector(job_description, vocabulary):
    binary_vector = []
    for skill in vocabulary:
        if skill in job_description["skills"]:
            binary_vector.append(1)
        else:
            binary_vector.append(0)
    return binary_vector

# ─────────────────────────────────────────────
# STEP 6 - COSINE SIMILARITY
# ─────────────────────────────────────────────
def compute_cosine_similarity(tfidf_vector, binary_vector):
    dot_product = sum(a * b for a, b in zip(tfidf_vector, binary_vector))
    magnitude_tfidf = math.sqrt(sum(a * a for a in tfidf_vector))
    magnitude_binary = math.sqrt(sum(a * a for a in binary_vector))
    if magnitude_tfidf == 0 or magnitude_binary == 0:
        return 0.0
    return dot_product / (magnitude_tfidf * magnitude_binary)

# ─────────────────────────────────────────────
# STEP 7 - RANK CANDIDATES
# ─────────────────────────────────────────────
def rank_candidates(resumes, job_description, tfidf_vectors, vocabulary):
    binary_vector = construct_binary_vector(job_description, vocabulary)
    scores = []
    for i, resume in enumerate(resumes):
        score = compute_cosine_similarity(tfidf_vectors[i], binary_vector)
        scores.append((resume["name"], score))
    # descending score, alphabetical on tie
    scores.sort(key=lambda x: (-x[1], x[0]))
    return scores[:3]

# ─────────────────────────────────────────────
# MAIN
# ─────────────────────────────────────────────
vocabulary = construct_vocabulary(resumes)
tfidf_vectors = compute_tfidf(resumes, vocabulary)

print("=" * 50)
print(f"Vocabulary ({len(vocabulary)} terms):")
print(vocabulary)

print("\n" + "=" * 50)
print("TF-IDF Vectors (non-zero only):")
for i, vector in enumerate(tfidf_vectors):
    non_zero = [(vocabulary[j], round(v, 4)) for j, v in enumerate(vector) if v > 0]
    print(f"{resumes[i]['name']:20s}: {non_zero}")

print("\n" + "=" * 50)
print("ALL CANDIDATES RANKED BY JD")
print("=" * 50)

for job_description in job_descriptions:
    binary_vector = construct_binary_vector(job_description, vocabulary)
    scores = []
    for i, resume in enumerate(resumes):
        score = compute_cosine_similarity(tfidf_vectors[i], binary_vector)
        scores.append((resume["name"], score))
    scores.sort(key=lambda x: (-x[1], x[0]))

    print(f"\n{job_description['id']} — {job_description['company']} ({job_description['role']})")
    print("-" * 40)
    for rank, (name, score) in enumerate(scores, 1):
        print(f"  {rank:2}. {name:20s} Score: {score:.4f}")

print("\n" + "=" * 50)
print("FINAL RESULTS")
print("=" * 50)
for job_description in job_descriptions:
    top3 = rank_candidates(resumes, job_description, tfidf_vectors, vocabulary)
    print(f"\n{job_description['id']} — {job_description['company']} ({job_description['role']})")
    print(", ".join(f"{name}({score:.2f})" for name, score in top3))
```
